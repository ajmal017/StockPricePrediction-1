```python
import pickle
import matplotlib
import os
os.getcwd()
```

```python
with open('../data/raw/stock_data.pickle', 'rb') as handle:
    ds = pickle.load(handle)
```

```python
dff = ds['tesla']
```

```python
import datetime as dt
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
from mpl_finance import candlestick_ohlc
import pandas as pd
```

```python
df = dff[['Date', 'Open', 'High', 'Low', 'Close']]
```

```python
df['Date'] = df['Date'].map(mdates.date2num)
```

```python
df = df[(df.index > 2200) & (df.index < 2300)]
```

```python
plt.figure(figsize=(20,10))
ax = plt.subplot()
candlestick_ohlc(ax, df.values, width=0.5, colorup='g', colordown='r')
ax.xaxis_date()
# ax.grid(True)
ax.set_axisbelow(True)
ax.set_title('TESLA - cena akcji')
ax.set_ylabel('Cena w USD')
plt.show()
```
